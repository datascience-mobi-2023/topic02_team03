### df of step 2 to 4

```python
# merge df_step_2_by and df_step_3_and_4_by
# create a new dictionary to store merged dataframes df_step_2_by and df_step_3_and_4_by
df_step_2_to_4_by = {}

# merge df_step_2_by[key_2] and respective df_step_3_and_4_by[key_3]
for key_2, df_step_2 in df_step_2_by.items():
    # split key_2 into muts_2
    muts_2 = key_2.split("_")
    for key_3, df_step_3_and_4 in df_step_3_and_4_by.items():
        # split key_3 into muts_3
        muts_3 = key_3.split("_")
        # compare muts_2 and muts_3, if they are the same, merge df_step_2 and df_step_3_and_4
        if muts_2[0] == muts_3[0] and muts_2[1] == muts_3[1]:
            df_step_2_to_4_by[key_2] = pd.concat([df_step_2, df_step_3_and_4])
        if muts_2[0] == muts_3[0] and muts_2[1] == muts_3[2]:
            df_step_2_to_4_by[key_2] = pd.concat([df_step_2, df_step_3_and_4])
        if muts_2[0] == muts_3[1] and muts_2[1] == muts_3[2]:
            df_step_2_to_4_by[key_2] = pd.concat([df_step_2, df_step_3_and_4])
        if muts_2[1] == muts_3[0] and muts_2[0] == muts_3[1]:
            df_step_2_to_4_by[key_2] = pd.concat([df_step_2, df_step_3_and_4])
        if muts_2[1] == muts_3[0] and muts_2[0] == muts_3[2]:
            df_step_2_to_4_by[key_2] = pd.concat([df_step_2, df_step_3_and_4])
        if muts_2[1] == muts_3[1] and muts_2[0] == muts_3[2]:
            df_step_2_to_4_by[key_2] = pd.concat([df_step_2, df_step_3_and_4])

# print df_step_2_to_4_by for each key_2
for key_2, df_step_2_to_4 in df_step_2_to_4_by.items():
    print(f"Filtered df for '{key_2}':")
    print(df_step_2_to_4)
    print()
```

### df of step 1 to 4

```python
# create a new dictionary to store merged dataframes df_step_2_to_4_by and df_1_mutation
df_step_1_to_4_by = {}

# add specific row from df_1_mutation to df_step_2_to_4_by if in this row contains a value of key_2
for key_2, df_step_2_to_4 in df_step_2_to_4_by.items():
    # split key_2 into muts_2
    muts_2 = key_2.split("_")
    # if row in df_step_1 contains a value of muts_2, add this row to df_step_2_to_4
    for index, row in df_1_mutation.iterrows():
        if row['m1'] in muts_2:
            df_step_1_to_4_by[key_2] = pd.concat([df_step_2_to_4, row.to_frame().T])

# print df_step_1_to_4_by for each key_2
for key_2, df_step_1_to_4 in df_step_1_to_4_by.items():
    print(f"Filtered df for '{key_2}':")
    print(df_step_1_to_4)
    print()
```

```python
# sort each df_step_1_to_4 by the number of mutations
# create a new dictionary to store sorted dataframes df_step_1_to_4_by
df_step_1_to_4_by_sorted = {}

# Create empty df_step_1_to_4_sorted for each key_2
for key_2, df_step_1_to_4 in df_step_1_to_4_by.items():
    df_step_1_to_4_sorted = pd.DataFrame(columns=['DMS_score', 'm1', 'm2', 'm3', 'm4'])
    df_step_1_to_4_by_sorted[key_2] = df_step_1_to_4_sorted

    # Add row of df_step_1_to_4 to df_step_1_to_4_sorted with None values in 'm2', 'm3', and 'm4'
    row_to_add = df_step_1_to_4[df_step_1_to_4['m2'].isnull() & df_step_1_to_4['m3'].isnull() & df_step_1_to_4['m4'].isnull()]
    df_step_1_to_4_sorted = df_step_1_to_4_sorted.append(row_to_add)
    # add row of df_step_1_to_4 to df_step_1_to_4_sorted with None values in 'm3' and 'm4' and not None value in 'm2'
    row_to_add = df_step_1_to_4[df_step_1_to_4['m2'].notnull() & df_step_1_to_4['m3'].isnull() & df_step_1_to_4['m4'].isnull()]
    df_step_1_to_4_sorted = df_step_1_to_4_sorted.append(row_to_add)
    # add row of df_step_1_to_4 to df_step_1_to_4_sorted with None value in 'm4' and not None value in 'm2' and 'm3'
    row_to_add = df_step_1_to_4[df_step_1_to_4['m2'].notnull() & df_step_1_to_4['m3'].notnull() & df_step_1_to_4['m4'].isnull()]
    df_step_1_to_4_sorted = df_step_1_to_4_sorted.append(row_to_add)
    # add row of df_step_1_to_4 to df_step_1_to_4_sorted with not None value in 'm2', 'm3', and 'm4'
    row_to_add = df_step_1_to_4[df_step_1_to_4['m2'].notnull() & df_step_1_to_4['m3'].notnull() & df_step_1_to_4['m4'].notnull()]
    df_step_1_to_4_sorted = df_step_1_to_4_sorted.append(row_to_add)

    # Add df_step_1_to_4_sorted to the dictionary
    df_step_1_to_4_by_sorted[key_2] = df_step_1_to_4_sorted

# Print df_step_1_to_4_by_sorted for each key_2
for key_2, df_step_1_to_4_sorted in df_step_1_to_4_by_sorted.items():
    print(f"Filtered df for '{key_2}':")
    print(df_step_1_to_4_sorted)
    print()
```
